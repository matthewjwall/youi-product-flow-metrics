```python
def get_issues(project):
    all_issues = []
    block_size = 100
    block_num = 0
    while True:
        start_idx = block_num*block_size
        issues = jira.search_issues('project=' + project, start_idx, block_size)
        if len(issues) == 0:
            # Retrieve issues until there are no more to come
            break
        block_num += 1
        all_issues.append(issues)
    all_issues_flat = [item for sublist in all_issues for item in sublist]
    return all_issues_flat

#rn = get_issues('RN')
#len(rn)
```
