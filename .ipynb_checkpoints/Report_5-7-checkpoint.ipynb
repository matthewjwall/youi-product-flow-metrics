```python
import pandas as pd
import datetime

pd.set_option('display.max_columns', None)

raw_count_data = pd.read_csv('PM_ENG_NRE_BUGS_5-7.csv')
raw_time_data = pd.read_csv('eng_changelog_clean.csv')

def get_key():
    time_data = []
    for key in raw_count_data['Issue key']:
        try:
            issue_time_rows = raw_time_data.loc[raw_time_data["eng_key"] == key]
            time_data.append(issue_time_rows)
        except:
            print("No time data: " + str(key))
    return pd.concat(time_data)
```

```python
times = get_key()
times
```

```python
from scripts import youi_utils

def changelog(single_issue, labels):
    cl_data = []
    issue = youi_utils.jira_auth().issue(single_issue, expand='changelog')
    print(issue)
    changelog = issue.changelog
    for history in changelog.histories:
        for item in history.items:
            if item.field == 'status':
                cl_item = [issue.key, issue.id, issue.raw['fields']['issuetype']['name'], history.created, item.toString, issue.raw['fields']['fixVersions'], labels]
                cl_item_df = pd.DataFrame([cl_item])
                #print(cl_item)
                cl_data.append(cl_item)
    return cl_data
                #cl_item_df.to_csv('eng_changelog.csv', mode='a', header=False)

count_unique = raw_count_data['Issue key'].unique()
benny_issue_times = []

for key in count_unique:
    issue_cl = changelog(key, ['N/A'])
    for item in issue_cl:
        benny_issue_times.append(item)
        print(item)
```

```python
final = pd.DataFrame.from_records(benny_issue_times, 
                                  columns=['issue_key', 'issue_id', 'issue_type', 'time', 'updated_status', 'fix_version', 'labels'])

final = final.sort_values(by=['issue_key'], ascending=False)

final.to_csv('./5-7_Issue_Changelog.csv', encoding='utf-8')



#benny_issue_times[0]

#final = pd.concat(benny_issue_times)
#pd.to_csv('5-7_Issue_Changelog.csv', encoding='utf-8')
```
