Notebook draft of JIRA project data ETL- gets project data and writes to a CSV.

```python
from jira import JIRA
import pandas as pd

user = 'matthew.wall@youi.tv'
apikey = 'TlbsBflmaTz8YTJ2WGDzCD20'
server = 'https://youilabs.atlassian.net'

options = {
 'server': server
}

jira = JIRA(options, basic_auth=(user,apikey) )
```

```python
def flatten_dict(d):
    def items():
        for key, value in d.items():
            if isinstance(value, dict):
                for subkey, subvalue in flatten_dict(value).items():
                    yield key + "." + subkey, subvalue
            else:
                yield key, value
    return dict(items())

#test = jira.search_issues('project = RN AND issuetype = bug AND status changed to closed AFTER -1w')
rn_throughput_raw = jira.search_issues('project = RN AND status changed to closed AFTER -1w')
rn_throughput_list = []
for item in rn_throughput_raw:
    flat_issue = flatten_dict(item.raw)
    rn_throughput_list.append(flat_issue)
```

```python
rn_throughput = pd.DataFrame.from_records(rn_throughput_list)
```
